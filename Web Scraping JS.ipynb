# Import all the files

```python
from bs4 import BeautifulSoup
import requests
import re
import html5lib
import pandas as pd
```

# We are going to take that sample HTML file and open it

```python
URL = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r = requests.get(URL)
```

# Make a Soup object of the site's HTML

```python
soup = BeautifulSoup(r.content, 'html5lib') #scrape the html
print(soup.prettify())
```

# Make a list of all soups who have the class ... so that all that is scraped are the products

```python
needed = soup.find_all('div', class_ = 'col-sm-4 col-lg-4 col-md-4')
print(needed)
```

# Find the first item of the list to base the scraping structure

```python
print(soup.find_all('div', class_ = 'col-sm-4 col-lg-4 col-md-4')[1])
```

# Create and store items into arrays of the pieces of HTML needed

```python
names = []
descriptions = []
prices = []
ratings = []

for container in needed:
    nameNotDone = container.find('a', attrs = {'class':'title'})
    name = nameNotDone['title']
    names.append(name)
    description = container.find('p', class_ = 'description').text
    descriptions.append(description)
    price = container.find('h4', class_ = 'pull-right price').text
    prices.append(price)
    rating = container.find('p', class_ = '')
    ratingNotDone = container.find('div', class_ = 'ratings')
    #print(len(ratingNotDone.findChildren()) - 2)
    ratings.append(len(ratingNotDone.findChildren()) - 2)
print(names)
print(descriptions)
print(prices)
print(ratings)
```

# Create a Panda DataFrame that makes the data more organized

```python
test_df = pd.DataFrame({'name': names,
'description': descriptions,
'price': prices,
'rating': ratings
})
print(test_df.info())
test_df
```
